```python
"""Bar Graph: Frequency of Earthquakes by Magnitude"""

# Group by magnitude and count occurrences
magnitude_counts = earthquake_data['mag'].value_counts().sort_index()
```

```python
# Creating a Bar Graph
plt.figure(figsize=(10, 6))
magnitude_counts.plot(kind='bar', color='green')
plt.title("Earthquakes' Frequencies by Magnitude")
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
plt.tight_layout()
```

```python
 """Seaborn is a library that uses Matplotlib underneath to plot graphs. 
 It will be used to visualize random distributions."""
 
 # importing the Seaborn 
 import seaborn as sns
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=earthquake_data['mag'], color='#5C8E3A')
plt.title('Box Plot of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.show()
plt.tight_layout()
```

```python
# Convert time column to datetime format
earthquake_data['time'] = pd.to_datetime(earthquake_data['time'])
```

```python
# Convert time column to datetime format
earthquake_data['time'] = pd.to_datetime(earthquake_data['time'])

# Group by date and count occurrences
earthquakes_by_day = earthquake_data.groupby(earthquake_data['time'].dt.date).size()

# Line Chart
plt.figure(figsize=(12, 6))
earthquakes_by_day.plot(kind='line', marker='o', color='purple')
plt.title('Number of Earthquakes Over Time (By Day)', fontsize=17, color='purple')
plt.xlabel('Date of Events', fontsize=15, color='#663300') # , font-weight: bold) #boldness 
plt.ylabel('Number of Earthquakes', fontsize=15, color = '#663300')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
plt.tight_layout()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(earthquake_data['depth'], earthquake_data['mag'], alpha=1.0, color='red')
plt.title('Scatter Plot of Earthquake Magnitude vs Depth')
plt.xlabel('Depth (km)')
plt.ylabel('Magnitude')
plt.show()
plt.tight_layout()
```

```python
plotdata = pd.DataFrame(
    {"latitude": [8821, 29, 7, 1, 85]}, 
    index=["earthquake", "explosive", "Ice Quake", "Other Events", "Quarry Blast"])
# Plot a bar chart
plotdata.plot(kind="bar", color='#e67e22')
plt.title('Types of Natural Disater', fontsize=14)
plt.xlabel('Type', fontsize=14, color='#21618c')
plt.ylabel('Latitude')
plt.tight_layout()
```

```python
labels = ['Califonia-USA', 'Philippines', 'Indonesia', 'Southwest Indian Ridge (SWIR)','Xizang-China']
sizes = [9, 3, 2, 1,2]  # approximately
explode = (0, 0.1, 0, 0,0)  # only "explode" the 2nd slice

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Places where Most Earthquakes Occured')
plt.show()
plt.tight_layout()
```
